```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def analyzer(df):
    
    df['DELIV_VALUE'] = ''
    df['5DDeVA'] = ''
    df['%Change_In_Price'] = ''
    df['%Change_In_Deliv_vs_AVG'] = ''
    
    df['DELIV_VALUE'] = (df['DELIV_QTY'] * df['AVG_PRICE'])/10000000
    
    for i in range(5):
        df['5DDeVA'][i] = np.nan
        df['%Change_In_Price'][i] = np.nan
        df['%Change_In_Deliv_vs_AVG'][i] = np.nan
        
    for i in range(5, df.shape[0]):

        df.at[i, '5DDeVA'] = df['DELIV_VALUE'].iloc[i-5:i].mean()
        df.at[i, '%Change_In_Price'] = ((df.at[i, 'CLOSE_PRICE'] - df.at[i-1, 'CLOSE_PRICE'])/df.at[i-1, 'CLOSE_PRICE']) * 100
        df.at[i, '%Change_In_Deliv_vs_AVG'] = (df.at[i, 'DELIV_VALUE']/df.at[i, '5DDeVA']) * 100
    
    return df
```

```python
df = pd.read_csv('SBIN.csv')
analyzer(df).head(20)
```
